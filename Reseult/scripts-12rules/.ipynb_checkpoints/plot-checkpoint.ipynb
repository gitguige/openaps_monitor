```python
import numpy as np 
import pandas as pd 
import os
from matplotlib import pyplot as plt
from scipy.spatial import distance
```

```python
# data_roc = pd.read_csv('summary_roc.csv')
data_roc = pd.read_csv('summary_roc_BGT.csv')
plt.plot(data_roc["FPR"],data_roc["TPR"],'g')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic curve')

plt.show()
```

```python
plt.plot(data_roc["FPR"],data_roc[" reaction_time"],'go',label='reaction_time')
plt.plot(data_roc["FPR"],data_roc["rectime_std"],'ro',label='std')

plt.xlabel('False Positive Rate')
plt.ylabel('Reaction time/minutes')
plt.legend(loc="center right")
plt.show()
```

```python
from sklearn.metrics import auc, roc_curve
tpr={}
fpr={}
AUC={}

tpr[0] = [0] + data_roc["TPR"].tolist() + [1.0]
fpr[0] = [0] + data_roc["FPR"].tolist() + [1.0]
AUC[0] = auc(fpr[0],tpr[0])

tpr[1]=[0,1,1]
fpr[1]=[0,0.2743,1]
AUC[1] = auc(fpr[1],tpr[1])

tpr[2]=[0,1,1]
fpr[2]=[0,0.1354,1]
AUC[2] = auc(fpr[2],tpr[2])

data_roc_wot = pd.read_csv('summary_roc_WOT.csv')
tpr[3] = [0] + data_roc_wot["TPR"].tolist() + [1.0]
fpr[3] = [0] + data_roc_wot["FPR"].tolist() + [1.0]
AUC[3] = auc(fpr[3],tpr[3])
plt.figure(figsize=(10,5))
# print(tpr)
# plt.plot(fpr[0],tpr[0],'r--',marker='o',label='ROC curve Context-aware with different $BGT$')# (area =%.2f)'%AUC[0])
plt.plot(fpr[3],tpr[3],'y--',marker='o',markersize=3,lw=2,label='ROC Curve: CAWOT')
plt.plot(fpr[0],tpr[0],'r--',marker='o',markersize=3,label='ROC Curve: CAWT')# (area =%.2f)'%AUC[0])
plt.plot(fpr[2][1],tpr[2][1],'g',marker='D',label='Chosen CAWT (BGT=(95,160)) (F1 =0.89)')
plt.plot(fpr[1][1],tpr[1][1],'b',marker='o',label='Baseline (F1 =0.77)')

plt.plot([0,1],[0,1],color='grey',lw=1,linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate', fontsize=15)
plt.ylabel('True Positive Rate', fontsize=15)
# plt.title('Receiver operating characteristic curve')
plt.legend(loc="best")
plt.show()
```

```python
distance.jaccard([0,1],[1,2])
hz_data = pd.read_csv("summary_monitor_hardware.csv")
```

```python
print(hz_data["alert"].tolist())
```

```python
import re
pattern = re.compile(r'\d+')
alert = hz_data["alert"].tolist()
alert_list=[]
for i in range(len(alert)-1):  
    result = pattern.findall(str(alert[i]))
    if result:
#         alert_list.append(result)
        alert_list += result
print(np.array(alert_list))
```

```python
unique_elements, counts_elements = np.unique(np.array(alert_list), return_counts=True)
```

```python
print("Frequency of unique values of the said array:")
print(np.asarray((unique_elements, counts_elements)))
```
