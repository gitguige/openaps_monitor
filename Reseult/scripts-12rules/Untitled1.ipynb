```python
import numpy as np
x=[(0,0), (1, -1), (2,-1.5), (3,2), (4,1.5), (5,1.5), (6, 0.8), (7,0), (8,-.5)]
h=[ (0,1), (1,-.8), (2,-.5), (3,-.2)]
```

```python
lenx =len(x)
lenh =len(h)
y=np.zeros(lenx+lenh-1)
for i in range(lenx):
    for j in range(lenh):
        y[i+j] += x[i][1]*h[j][1]
```

```python
y1=[]
for i in range(len(y)):
    y1.append((i,y[i]))

print(y1)
```

```python
from matplotlib import pyplot as plt
x1=[(0,0), (1,80), (2, 85), (3, 90) ,(4,87) ,(5, 88)]
lenx =len(x1)
for i in range(lenx):
    x= range(lenx)
    y = np.zeros(lenx) #all zeros
    y[i] = x1[i][1] #replace one item
    plt.plot(x,y,'go-')
    plt.ylabel('X%i[n]'%i, rotation=0)
    
    plt.show()
```
